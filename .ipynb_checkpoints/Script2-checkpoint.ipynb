```python
import pandas as pd
import re
from collections import Counter

# --- Configuración Inicial ---

# Nombre del archivo de Excel confirmado por tu listado de directorio.
nombre_archivo_excel = 'tablabase_plana3.xlsx' 

try:
    # Usamos pd.read_excel() para leer el archivo de Excel.
    df = pd.read_excel(nombre_archivo_excel)
    print(f"¡Éxito! El archivo '{nombre_archivo_excel}' se cargó correctamente.")
    
    # --- PROCESO PRINCIPAL DE ANÁLISIS ---
    # (El resto del código se ejecutará automáticamente si la carga fue exitosa)

    # Lista de stopwords en español
    stopwords_es = [
        'de', 'la', 'que', 'el', 'en', 'y', 'a', 'los', 'del', 'se', 'las', 'por', 'un',
        'para', 'con', 'no', 'una', 'su', 'al', 'lo', 'como', 'más', 'pero', 'sus',
        'le', 'ya', 'o', 'este', 'ha', 'sí', 'porque', 'esta', 'son', 'entre',
        'está', 'cuando', 'muy', 'sin', 'sobre', 'también', 'me', 'hasta', 'hay',
        'donde', 'quien', 'desde', 'todo', 'nos', 'durante', 'todos', 'uno',
        'les', 'ni', 'contra', 'otros', 'ese', 'eso', 'ante', 'ellos', 'e', 'esto',
        'mí', 'antes', 'algunos', 'qué', 'ninguno', 'mi', 'tu', 'te', 'ti', 'tú',
        'nosotros', 'vosotros', 'ellos', 'ellas', 'es', 'era', 'eres', 'soy', 'somos',
        'sois', 'eran', 'fui', 'fue', 'fuimos', 'fueron', 'ser', 'siendo', 'sido'
    ]

    # Funciones de Análisis
    def get_top_ngrams(text, n=1, top_k=10):
        words = re.findall(r'\b\w+\b', text.lower())
        words = [word for word in words if word not in stopwords_es and not word.isdigit()]
        if n == 1: ngrams = words
        else: ngrams = zip(*[words[i:] for i in range(n)])
        return Counter(ngrams).most_common(top_k)

    categories = df['Bloque temático'].unique()
    for category in categories:
        if pd.isna(category): continue
        print("\n" + "="*80)
        print(f"### CATEGORÍA: {category.strip()}")
        print("="*80)
        category_df = df[df['Bloque temático'] == category].copy()
        top_sectors = category_df['employment sector'].value_counts().nlargest(3).index.tolist()
        reason_df = category_df.dropna(subset=['reason']).copy()
        reason_df['reason_length'] = reason_df['reason'].str.len()
        
        print("\n--- 1. Análisis Cuantitativo de la Argumentación ---\n")
        if reason_df.empty: print("No se encontraron argumentos en esta categoría.")
        else:
            stats = reason_df[reason_df['employment sector'].isin(top_sectors)].groupby('employment sector')['reason_length'].agg(['count', 'mean']).round(2)
            stats.columns = ['Nº de Argumentos', 'Longitud Promedio (caracteres)']
            if stats.empty: print("No hay argumentos de los 3 sectores principales para analizar.")
            else: print(stats)
        
        print("\n--- 2. Análisis Cualitativo: Conceptos Clave por Sector ---\n")
        if reason_df.empty: print("No hay argumentos para extraer conceptos clave.")
        else:
            for sector in top_sectors:
                print(f"\n**Sector: {sector}**")
                sector_reasons_df = reason_df[reason_df['employment sector'] == sector]
                if sector_reasons_df.empty:
                    print("  No se encontraron argumentos para este sector.")
                    continue
                full_text = ' '.join(sector_reasons_df['reason'])
                top_unigrams = get_top_ngrams(full_text, n=1, top_k=10)
                top_bigrams = get_top_ngrams(full_text, n=2, top_k=5)
                unigram_str = ', '.join([f"'{word[0]}'" for word, count in top_unigrams])
                bigram_str = ', '.join([f"'{' '.join(words)}'" for words, count in top_bigrams])
                print(f"  - Palabras Clave (Unigramas): {unigram_str}")
                print(f"  - Conceptos Clave (Bigramas): {bigram_str}")
    print("\n" + "="*80)
    print("Análisis completado.")
    print("="*80)

except FileNotFoundError:
    print(f"Error: El archivo '{nombre_archivo_excel}' sigue sin encontrarse.")

except Exception as e:
    print(f"Ocurrió un error inesperado al leer el archivo: {e}")
    print("\nSUGERENCIA: Si el error menciona 'openpyxl', necesitas instalar una librería adicional.")
    print("Ejecuta el siguiente comando en una nueva celda de tu notebook y luego vuelve a correr esta celda:")
    print("!pip install openpyxl")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def generar_grafico_percepciones(dataframe, nombre_categoria):
    try:
        category_df = dataframe[dataframe['Bloque temático'] == nombre_categoria].copy()
        if category_df.empty:
            print(f"-> Advertencia: No se encontraron datos para la categoría '{nombre_categoria}'. Saltando...")
            return

        category_df['employment sector'] = category_df['employment sector'].str.replace('Ciencia y TecnologÃ­a; Science', 'Ciencia y Tecnología; Science', regex=False)
        top_sectors = category_df['employment sector'].value_counts().nlargest(3).index.tolist()
        analysis_df = category_df[category_df['employment sector'].isin(top_sectors)]

        if analysis_df.empty:
            print(f"-> Advertencia: No hay suficientes datos de los sectores principales para graficar '{nombre_categoria}'. Saltando...")
            return

        perception_by_sector = analysis_df.groupby('employment sector')[['int_ans1', 'int_ans2', 'int_ans3']].mean()
        perception_by_sector.columns = ['Probabilidad', 'Impacto', 'Capacidad de Respuesta']

        short_category_name = nombre_categoria.split(':')[0].strip()
        title = f'Percepciones Promedio por Sector - {short_category_name}'
        
        ax = perception_by_sector.plot(kind='bar', figsize=(12, 7), width=0.8, edgecolor='black', zorder=3)
        plt.title(title, fontsize=16, pad=20)
        plt.xlabel('Sector Laboral', fontsize=12)
        plt.ylabel('Valoración Promedio (Escala 1-10)', fontsize=12)
        plt.xticks(rotation=0, ha='center', fontsize=10)
        plt.grid(axis='y', linestyle='--', alpha=0.7, zorder=0)
        plt.legend(title='Variables de Percepción', fontsize=10)
        plt.tight_layout()

        clean_filename = "".join([c for c in short_category_name if c.isalpha() or c.isdigit() or c==' ']).rstrip().replace(" ", "_")
        output_filename = f'grafico_percepciones_{clean_filename}.png'
        plt.savefig(output_filename)
        
        # Al no cerrar la figura con plt.close(), Jupyter la mostrará automáticamente.
        plt.show() 
        
        print(f"✓ Gráfico generado para '{short_category_name}' y guardado como '{output_filename}'.")

    except Exception as e:
        print(f"✗ Ocurrió un error al generar el gráfico para '{nombre_categoria}': {e}")

nombre_archivo_excel = 'tablabase_plana3.xlsx'
try:
    df = pd.read_excel(nombre_archivo_excel)
    print(f"Archivo '{nombre_archivo_excel}' cargado exitosamente.\n")
    
    print("Iniciando la generación de gráficos para todas las categorías...")
    
    df.columns = df.columns.str.strip()
    
    categories = df['Bloque temático'].dropna().unique()
    for category in categories:
        generar_grafico_percepciones(df, category)
        
    print("\nProceso completado. Todos los gráficos han sido generados y mostrados.")

except FileNotFoundError:
    print(f"Error Crítico: No se encontró el archivo '{nombre_archivo_excel}'. Verifica el nombre y la ubicación.")
except Exception as e:
    print(f"Ocurrió un error inesperado durante la carga del archivo: {e}")
```
